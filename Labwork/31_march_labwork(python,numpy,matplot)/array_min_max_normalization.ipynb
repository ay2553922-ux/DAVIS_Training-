```python
"""
Problem 16: Normalize array values between 0 and 1
"""

import numpy as np

def get_input(size=6):
    """
    Collect array input from user
    """
    data = []
    print(f"Enter at least {size} numbers:")

    while len(data) < size:
        try:
            data.extend(map(int, input("Input: ").split()))
        except ValueError:
            print("Invalid input.")

    return np.array(data[:size])


# Take input array
arr = get_input()

# Find minimum and maximum values
min_val = np.min(arr)
max_val = np.max(arr)

# Normalize using formula (x - min) / (max - min)
normalized = (arr - min_val) / (max_val - min_val)

# Display result
print("\nNormalized Array:", normalized)
```
