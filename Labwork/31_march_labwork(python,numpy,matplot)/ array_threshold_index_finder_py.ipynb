```python
"""
Problem 18: Find indices of elements greater than a given value
"""

import numpy as np

def get_input(size=20):
    """
    Collect array input from user
    """
    data = []
    print(f"Enter at least {size} numbers:")

    while len(data) < size:
        try:
            data.extend(map(int, input("Input: ").split()))
        except ValueError:
            print("Invalid input.")

    return np.array(data[:size])


# Take input array
arr = get_input()

# Take comparison value
value = int(input("Enter value: "))

# Find indices where condition is satisfied
indices = np.where(arr > value)

# Display result
print("\nIndices:", indices)
```
