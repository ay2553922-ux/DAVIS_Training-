```python
import matplotlib.pyplot as plt

# Input data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

# Create line plot
plt.plot(x, y, marker='o')

# Add title and labels
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Input data (5 products)
products = ["Product A", "Product B", "Product C", "Product D", "Product E"]
sales = [150, 200, 120, 300, 250]

# Create bar chart
plt.bar(products, sales)

# Add title and labels
plt.title("Product Sales Bar Chart")
plt.xlabel("Products")
plt.ylabel("Sales")

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Input data (category-wise)
categories = ["Food", "Rent", "Transport", "Entertainment", "Others"]
values = [300, 500, 150, 100, 50]

# Create pie chart
plt.pie(values, labels=categories, autopct='%1.1f%%')

# Add title
plt.title("Expense Distribution")

# Show chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random dataset
data = np.random.randint(1, 100, 50)  # 50 random numbers between 1 and 100

# Create histogram
plt.hist(data, bins=10)

# Add title and labels
plt.title("Histogram of Random Data")
plt.xlabel("Value Range")
plt.ylabel("Frequency")

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Input data
x = [1, 2, 3, 4, 5]

# Two datasets
y1 = [10, 20, 15, 25, 30]
y2 = [5, 15, 10, 20, 25]

# Plot both datasets
plt.plot(x, y1, marker='o', label="Dataset 1")
plt.plot(x, y2, marker='s', label="Dataset 2")

# Add title and labels
plt.title("Comparison of Two Datasets")
plt.xlabel("X-axis")
plt.ylabel("Values")

# Show legend
plt.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Input data
x = [1, 2, 3, 4, 5]
y = [10, 25, 15, 30, 20]

# Create scatter plot
plt.scatter(x, y)

# Find maximum point
max_y = max(y)
max_index = y.index(max_y)
max_x = x[max_index]

# Highlight maximum point
plt.scatter(max_x, max_y, s=100, marker='o', label="Maximum Point")

# Add title and labels
plt.title("Scatter Plot with Maximum Point Highlighted")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add grid
plt.grid()

# Show legend
plt.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Input data (monthly revenue)
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
revenue = [2000, 2500, 2200, 3000, 2800, 3500]

# Create line plot
plt.plot(months, revenue, marker='o')

# Add title and labels
plt.title("Monthly Revenue Analysis")
plt.xlabel("Months")
plt.ylabel("Revenue (₹)")

# Add grid
plt.grid(True, linestyle='--')

# Show plot
plt.show()
```
