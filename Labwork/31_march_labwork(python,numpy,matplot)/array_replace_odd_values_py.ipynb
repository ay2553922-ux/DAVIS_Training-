```python
"""
Problem 14: Replace all odd numbers in an array with -1
"""

import numpy as np

def get_input(size=20):
    """
    Collect array input from user
    """
    data = []
    print(f"Enter52 numbers:")

    while len(data) < size:
        try:
            data.extend(map(int, input("Input: ").split()))
        except ValueError:
            print("Invalid input.")

    return np.array(data[:size])


# Take input array
arr = get_input()

# Replace odd numbers using boolean indexing
arr[arr % 2 != 0] = -1

# Display result
print("\nModified Array:", arr)
```
