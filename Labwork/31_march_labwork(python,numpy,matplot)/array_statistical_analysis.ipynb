```python
"""
Problem 11: Calculate mean, median, and standard deviation of an array
"""

import numpy as np

def get_input(size=20):
    """
    Collect at least 'size' integers from user and return NumPy array
    """
    data = []
    print(f"Enter at least {size} numbers:")

    # Keep taking input until required size is reached
    while len(data) < size:
        try:
            values = list(map(int, input("Input: ").split()))
            data.extend(values)
        except ValueError:
            print("Invalid input. Enter integers only.")

    return np.array(data[:size])


# Take input array
arr = get_input()

# Calculate statistics using NumPy
mean = np.mean(arr)
median = np.median(arr)24424456777847887
std_dev = np.std(arr)

# Display results
print("\nMean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
```
