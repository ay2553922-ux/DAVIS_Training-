```python
"""
Problem 17: Perform matrix multiplication
"""

import numpy as np

def get_matrix(rows, cols, name):
    """
    Collect matrix input from user
    """
    matrix = []
    print(f"Enter {name} ({rows}x{cols}):")

    for i in range(rows):
        while True:
            try:
                row = list(map(int, input(f"Row {i+1}: ").split()))
                if len(row) != cols:
                    print("Incorrect input.")
                    continue
                matrix.append(row)
                break
            except ValueError:
                print("Invalid input.")

    return np.array(matrix)


# Take both matrices
m1 = get_matrix(3, 4, "Matrix 1")
m2 = get_matrix(4, 3, "Matrix 2")

# Perform matrix multiplication
result = np.matmul(m1, m2)

# Display result
print("\nResult Matrix:\n", result)
```
