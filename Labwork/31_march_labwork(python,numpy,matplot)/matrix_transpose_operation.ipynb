```python
"""
Problem 13: Transpose a 2D matrix
"""

import numpy as np

def get_matrix(rows=4, cols=5):
    """
    Collect matrix input from user
    """
    matrix = []
    print(f"Enter {rows}x{cols} matrix:")

    for i in range(rows):
        while True:
            try:
                row = list(map(int, input(f"Row {i+1}: ").split()))
                if len(row) != cols:
                    print("Enter correct number of elements.")
                    continue
                matrix.append(row)
                break
            except ValueError:
                print("Invalid input.")

    return np.array(matrix)


# Take matrix input
matrix = get_matrix()

# Transpose matrix using NumPy
transpose = matrix.T

# Display result
print("\nTransposed Matrix:\n", transpose)
```
