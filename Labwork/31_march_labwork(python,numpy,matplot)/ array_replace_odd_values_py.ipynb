```python
"""
Problem 15: Extract diagonal elements from a matrix
"""

import numpy as np

def get_matrix(size=5):
    """
    Collect square matrix input from user
    """
    matrix = []
    print(f"Enter {size}x{size} matrix:")

    for i in range(size):
        while True:
            try:
                row = list(map(int, input(f"Row {i+1}: ").split()))
                if len(row) != size:
                    print("Incorrect input.")
                    continue
                matrix.append(row)
                break
            except ValueError:
                print("Invalid input.")

    return np.array(matrix)


# Take matrix input
matrix = get_matrix()

# Extract diagonal elements
diagonal = np.diag(matrix)

# Display result
print("\nDiagonal Elements:", diagonal)
```
