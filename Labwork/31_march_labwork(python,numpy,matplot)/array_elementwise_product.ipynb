```python
"""
Problem 12: Perform element-wise multiplication of two arrays
"""

import numpy as np

def get_input(size=20, name="array"):
    """
    Collect array input from user
    """
    data = []
    print(f"Enter at least {size} elements for {name}:")

    while len(data) < size:
        try:
            data.extend(map(int, input("Input: ").split()))
        except ValueError:
            print("Invalid input.")

    return np.array(data[:size])


# Take both arrays as input
arr1 = get_input(name="Array 1")
arr2 = get_input(name="Array 2")

# Perform element-wise multiplication
result = arr1 * arr2

# Display result
print("\nResult:", result)3
```
